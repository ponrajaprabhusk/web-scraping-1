# Top Repositories for Github Topics

### Outline:
- We're going to scrape https://github.com/topics
- We'll get a list of topics. For each topic, we'll get topic title, topic page URL and topic description
- For each topic, we will get the top 25 repositories in the topic from the topic page
- For each repo, we'll grab the repo name, username, stars and repo URL
- For each topic we'll create a CSV file in the folln format:
```
    Repo Name,Username,Stars,Repo URL
    three.js,mrdoob,69700,https://github.com/mrdoob/three.js

```

````python
import requests
````

````python
topics_Url = 'https://github.com/topics'
````

````python
response = requests.get(topics_Url)
````

````python
response.status_code
````

````python
len(response.text)
````

````python
page_contents = response.text 
````

````python
page_contents[:1000]
````

````python
with open('webpage.html', 'w', encoding="utf-8") as f:
    f.write(page_contents)
````

````python
!pip install beautifulsoup4 --quiet
````

````python
from bs4 import BeautifulSoup
````

````python
doc = BeautifulSoup(page_contents,'html.parser')
````

````python
doc
````

````python
p_tags = doc.findAll('p')
````

````python
len(p_tags)
````

````python
p_tags[:5]
````

````python
selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'

topic_title_tags = doc.find_all('p',{'class': selection_class})

# or for class we can also write class_=selection_class
````

````python
len(topic_title_tags)
````

````python
topic_title_tags
````

````python
topic_title_tags[:5]
````

````python
desc_selector = 'f5 color-fg-muted mb-0 mt-1'
topic_desc_tags = doc.findAll('p',{'class': desc_selector})
````

````python
topic_desc_tags[:5]
````

````python
topic_link_tags = doc.findAll('a',{'class':'no-underline flex-1 d-flex flex-column'})
````

````python
len(topic_link_tags)
````

````python
topic0_url = "https://github.com" +  topic_link_tags[0]['href']
````

````python
topic0_url
````

````python
topic_title_tags[0].text
````

````python
topic_titles = []

for tag in topic_title_tags:
    topic_titles.append(tag.text)

print(topic_titles)    
````

````python
topic_descriptions = []

for tag in topic_desc_tags:
    topic_descriptions.append(tag.text.strip())

print(topic_descriptions[:5])    
````

````python
topic_urls = []
base_url = 'https://github.com'

for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])

topic_urls
````

````python
!pip install pandas --quiet
````

````python
import pandas as pd 
````

````python
topics_dict ={
    'title':topic_titles,
    'description': topic_descriptions,
    'url': topic_urls
}
````

````python
topics_df = pd.DataFrame(topics_dict)
````

````python
topics_df
````

````python
topics_df.to_csv('topics.csv', index= None)
````
